```python
import matplotlib.pyplot as plt

# 1. Inspect current columns
print("Columns in df:", df.columns.tolist())

# 2. Create AttritionFlag if missing
if 'AttritionFlag' not in df.columns:
    df['AttritionFlag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
    print("Added AttritionFlag.  Sample values:\n", df[['Attrition','AttritionFlag']].head())

# 3. Now compute and plot
dept_attr = df.groupby('Department')['AttritionFlag'].mean() * 100

fig, ax = plt.subplots(figsize=(8,4))
dept_attr.sort_values(ascending=False).plot(
    kind='bar', 
    color='salmon', 
    ax=ax
)

# Labels, title, annotations
ax.set_ylabel('Attrition Rate (%)')
ax.set_title('Attrition % by Department')
for i, pct in enumerate(dept_attr.sort_values(ascending=False)):
    ax.text(i, pct + 0.5, f"{pct:.1f}%", ha='center')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# 7.1.1 Select features
features = ['Age','DistanceFromHome','MonthlyIncome','NumCompaniesWorked',
            'PercentSalaryHike','TotalWorkingYears','YearsAtCompany','OverTime']
X = df[features].copy()

# 7.1.2 Encode Overtime and scale
X['OverTimeFlag'] = (X['OverTime']=='Yes').astype(int)
X.drop(columns='OverTime', inplace=True)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

y = df['AttritionFlag']

# 7.1.3 Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_prob))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

importances = pd.Series(rf.feature_importances_, index=X.columns)
importances.sort_values().plot(kind='barh', figsize=(6,4), color='teal')
plt.title('Feature Importance')
plt.show()
```

```python


#Ensure AttritionFlag exists
if 'AttritionFlag' not in df.columns:
    df['AttritionFlag'] = df['Attrition'].map({'Yes':1, 'No':0})


def plot_attrition_by(col, figsize=(6,4), palette='viridis', order=None):
    rates = df.groupby(col)['AttritionFlag'].mean().mul(100)
    if order:
        rates = rates.reindex(order)
    fig, ax = plt.subplots(figsize=figsize)
    rates.plot(kind='bar', color=sns.color_palette(palette, len(rates)), ax=ax)
    ax.set_ylabel('Attrition Rate (%)')
    ax.set_title(f'Attrition % by {col}')
    ax.set_ylim(0, rates.max() + 5)
    for i, pct in enumerate(rates):
        ax.text(i, pct + 0.5, f"{pct:.1f}%", ha='center')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()





```

```python
#Gender
plot_attrition_by('Gender', palette='pastel')
```

```python
#Marital Status
plot_attrition_by('MaritalStatus', palette='Set2', order=['Single','Married','Divorced'])


```

```python


#Education Field
plot_attrition_by('EducationField', palette='cubehelix')


```

```python


#Business Travel
plot_attrition_by('BusinessTravel', palette='magma', 
                  order=['Non-Travel','Travel_Rarely','Travel_Frequently'])

```
